```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Show plots in notebook (if you're in Jupyter)
%matplotlib inline

# Load cleaned HR data
df = pd.read_csv(r"D:\projects_github\New folder\final_cleaned_hr_data.csv")

# Display first few rows
df.head()

# --------------------------------------------
# 1. Attrition Count
plt.figure(figsize=(6, 4))
sns.countplot(x='Attrition', data=df, palette='Set2')
plt.title("Attrition Count")
plt.xlabel("Attrition")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
plt.savefig("attrition_count.png")

# --------------------------------------------
# 2. Create Age Groups
bins = [18, 25, 35, 45, 55, 100]
labels = ['<25', '25-34', '35-44', '45-54', '55+']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# 3. Attrition Rate by Age Group (%)
age_attrition = df.groupby('Age_Group')['Attrition'].value_counts(normalize=True).unstack().fillna(0) * 100
age_attrition.plot(kind='bar', stacked=True, colormap='coolwarm', figsize=(8, 5))
plt.title("Attrition Rate by Age Group (%)")
plt.ylabel("Percentage")
plt.xlabel("Age Group")
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()
plt.savefig("age_group_attrition.png")

# --------------------------------------------
# 4. Attrition by Gender
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', hue='Attrition', data=df, palette='husl')
plt.title("Attrition by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Employees")
plt.tight_layout()
plt.show()
plt.savefig("gender_attrition.png")

# --------------------------------------------
# 5. Attrition by Department
plt.figure(figsize=(7, 4))
sns.countplot(x='Department', hue='Attrition', data=df, palette='Set3')
plt.title("Attrition by Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
plt.savefig("department_attrition.png")


# --------------------------------------------
# 6. Attrition by Job Role
plt.figure(figsize=(10, 5))
sns.countplot(x='JobRole', hue='Attrition', data=df, palette='Paired')
plt.title("Attrition by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Number of Employees")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig("jobrole_attrition.png")
```
